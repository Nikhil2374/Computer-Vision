```python

#Gorav_21BCS1852

import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('sample.png')

# Convert to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to grayscale using formula
gray_image_formula = np.dot(rgb_image[...,:3], [0.299, 0.587, 0.114])

# Convert to grayscale using OpenCV function
gray_image_cv2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the images
plt.figure(figsize=(12, 8))

plt.subplot(1,3,1)
plt.imshow(rgb_image) -[]

plt.title('RGB')

plt.subplot(1,3,2)
plt.imshow(gray_image_formula, cmap='gray')
plt.title('Grayscale (Formula)')

plt.subplot(1,3,3)
plt.imshow(gray_image_cv2, cmap='gray')
plt.title('Grayscale (OpenCV)')

plt.show()
```
